```python
# Load libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Designate the path where I saved the datasets

# YOU CAN CHANGE THE PATH TO YOUR FILE LOCATION

# Load the data using pandas read_csv function. 
data = pd.read_csv("used_in_oprobitregression_data.csv")
data.shape  # This must give you (4677, 52), otherwise you have not
            # generated the file from the code in the do file or you
            # are loading the wrong csv file.
```

```python
# Specify a valid file path
save_path = os.path.join(os.getcwd(), 'figure_1.tif')  # Save in the current working directory

# Set Seaborn style
sns.set_theme(style="ticks")  # Change to "white" or other styles if preferred

# Define custom palette
custom_palette = {'Not Prev EV': 'grey', 'Prev EV': 'green'}

# Create a new column with descriptive labels
data['prev_evs_label'] = data['prev_evs'].map({0: 'Not Prev EV', 1: 'Prev EV'})

# Set font to Times New Roman and customize font sizes
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': 'Times New Roman',
    'axes.titlesize': 18,
    'axes.labelsize': 18,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18,
})

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14.17, 10.67))  # Convert width and height from pixels to inches

# Create KDE plot
sns.kdeplot(data=data, x="attitude_reduce_ghge_fix", hue="prev_evs_label", fill=True, palette=custom_palette, ax=ax, linewidth=3)

# Remove legend title
leg = ax.get_legend()
if leg is not None:
    leg.set_title(None)
    # Set font size directly
    for label in leg.get_texts():
        label.set_fontsize(32)  # Set the font size of the legend labels
    # Set the legend position
    # Adjust bbox_to_anchor values to position the legend
    leg.set_bbox_to_anchor((0.3, 0.98))  # Adjust this tuple to fine-tune the position

    # Increase the size of the color boxes (handles)
    for handle in leg.legend_handles:  # Iterate through all legend handles
        handle.set_width(20)  # Adjust width of the color box (marker)
        handle.set_height(20)  # Adjust height of the color box (marker)

# Customize axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

# Customize ticks
ax.tick_params(axis='both', which='major', direction='out', length=10, width=1.5)
ax.tick_params(axis='both', which='minor', direction='out', length=5, width=1.5)

# Customize grid
ax.grid(True, linestyle='--', linewidth=1, color='#f0f0f0')
    
# Add custom labels below the x-axis ticks
xticks = [-3, -2, -1, 0, 1, 2, 3]
custom_labels = {
    -3: 'Not at all important',
    3: 'Extremely important'
}

# Set x-axis ticks and ensure default labels are shown
ax.set_xticks(xticks)

# Add custom labels above each tick
for tick, label in custom_labels.items():
    # Add text annotation above each tick
    ax.text(tick, ax.get_ylim()[1] + -0.382, label, ha='center', va='bottom', fontsize=34)
# Customize plot
plt.title("\nDensity plot of the attitudes towards reducing\nGHG emissions by previous ownership of EV", fontsize=40)
plt.xlabel("\nImportance of reducing GHG emissions", fontsize=36)
plt.ylabel("Density", fontsize=36)
plt.xticks(fontsize=36)
plt.yticks(fontsize=36)

# Save the plot in TIF format with 600 DPI
plt.savefig(save_path, format='tiff', dpi=600, bbox_inches='tight', pad_inches=0.1)

# Show plot
plt.show()
```
